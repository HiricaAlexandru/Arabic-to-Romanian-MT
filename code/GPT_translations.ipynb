```python
from openai import OpenAI
from google.colab import userdata

client = OpenAI()

def run_inference_with_gpt4(prompt):
  completion = client.chat.completions.create(
      model="gpt-4o",
      messages=[
          {"role": "developer", "content": "You are a helpful assistant."},
          {
              "role": "user",
              "content": f"Translate the following arabic text: '{prompt}' into Romanian. Write only the translation and nothing else."
          }
      ]
  )
  return completion.choices[0].message.content
```

```python
import os
import shutil

def reflect_folder_structure(input_folder, output_folder):
    """
    Replicates the folder structure of the input folder into the output folder.

    Args:
        input_folder (str): Path to the input folder.
        output_folder (str): Path to the output folder.
    """
    for root, dirs, _ in os.walk(input_folder):
        # Get the relative path of the current folder from the input folder
        relative_path = os.path.relpath(root, input_folder)
        # Create the corresponding folder in the output folder
        output_path = os.path.join(output_folder, relative_path)
        os.makedirs(output_path, exist_ok=True)

# Example usage
input_folder_path = "/content/drive/MyDrive/disertatie/translate/OPENNMT/GEMINI"
output_folder_path = "/content/drive/MyDrive/disertatie/translate/GPT4/"

reflect_folder_structure(input_folder_path, output_folder_path)
print(f"Folder structure replicated from '{input_folder_path}' to '{output_folder_path}'.")
```

```python
from tqdm import tqdm

def read_file_line_by_line(input_file):
    with open(input_file, "r") as f:
        lines = f.readlines()
    return lines

def write_to_file(output_file, lines):
    with open(output_file, "w") as f:
      for line in lines:
        f.write(f"{line}\n")

def create_arabic_romanian_translations(all_lines):
  translations = []
  for idx, line in enumerate(tqdm(all_lines)):
    translation = run_inference_with_gpt4(line)
    translations.append(translation)
  return translations
```

```python
INPUT_FOLDER = "/content/drive/MyDrive/disertatie/dataset/"
OUTPUT_FOLDER = "/content/drive/MyDrive/disertatie/translate/GPT4/"
```

```python
INPUT_FILE = f"{INPUT_FOLDER}from_training_dataset/ar.txt"
lines_input = read_file_line_by_line(INPUT_FILE)
translations = create_arabic_romanian_translations(lines_input)
write_to_file(f"{OUTPUT_FOLDER}from_training_dataset/gpt4.txt", translations)
```
