```python
import os,json
import os
import numpy as np

os.environ["WANDB_DISABLED"] = "true"


! mkdir -p metx/jsons

def do_matrix(dataset, model):
  ref_path = f'{PATH_TO_APPEND}ar-ro/dataset/{dataset}/ro.txt'
  outfis = f'metx/scores/{model}%{dataset}.jsonl'
  if os.path.exists(outfis):
    return
  print(model)
  hyp_file_name = meta[model]
  if model == "OPENNMT_BASELINE":
    hyp_file_name = "OPEN_NMT.txt"

  hyp_path = f'{PATH_TO_APPEND}ar-ro/translate/{model}/{dataset}/{hyp_file_name}'
  if model == "human":
    hyp_path = ref_path
  print("PATH: ", hyp_path, ref_path)
  reference = read_fis(ref_path)
  hypothesis = read_fis(hyp_path)
  json_lines = []
  for l in range(len(hypothesis)):
      r = reference[l].strip()
      h = hypothesis[l].strip()
      json_lines.append({"reference":r,"hypothesis":h})

  if "twitter" in dataset:
    dataset = "twitter"

  json_file = f'metx/jsons/{model}%{dataset}.jsonl'
  with open(json_file, 'w') as ho:
      json.dump(json_lines, ho, indent=4)
  ! mkdir -p metx/scores/
  outfis = f'metx/scores/{model}%{dataset}.jsonl'
  !python metricx/predict.py \
    --tokenizer google/mt5-xl \
    --model_name_or_path google/metricx-23-xl-v2p0 \
    --max_input_length 1024 \
    --batch_size 1 \
    --input_file {json_file} \
    --output_file {outfis}

def eval_matrix(dataset, model):

  if "twitter" in dataset:
    dataset = "twitter"
  outfis = f'{PATH_TO_APPEND}/metrics/metx/scores/{model}%{dataset}.jsonl'

  sscores = []
  if not os.path.exists(outfis):
    print('missing ', outfis)
    return -1
  with open(outfis) as seg_scores:
    for s in seg_scores:
      sscores.append(json.loads(s)['prediction'])
  print(model, ' ', np.mean(sscores))
  return np.mean(sscores)


def eval_matrix_multi(multiple_datasets, model):
  sscores = []
  for dataset in multiple_datasets:
    if "twitter" in dataset:
      dataset = "twitter"
    outfis = f'{PATH_TO_APPEND}/metrics/metx/scores/{model}%{dataset}.jsonl'
    if not os.path.exists(outfis):
      print('missing ', outfis)
      return -1
    with open(outfis) as seg_scores:
      for s in seg_scores:
        sscores.append(json.loads(s)['prediction'])
    print(model, ' ', dataset, ' ', np.mean(sscores))
  return np.mean(sscores)
```

```python
#import pandas as pd
#df = pd.read_csv('comet.scores', sep='\t', header=None)
#df[2].apply(lambda x: x.split(':'))

def parse_comet(fis_in):
  with open(fis_in) as sc:
    for line in sc:
      if "Segment" not in line:
        score = line.strip().split()[-1]
        return score

```

```python
def do_comet(dataset, model):
  ! mkdir -p comet
  ref_path = f'{PATH_TO_APPEND}ar-ro/dataset/{dataset}/ro.txt'
  src_path = f'{PATH_TO_APPEND}ar-ro/dataset/{dataset}/ar.txt'

  prev_dataset = dataset
  if "twitter" in dataset:
    dataset = "twitter"

  out_fis = f'comet/{model}%{dataset}.scores'
  print(dataset, ' ', model)
  if os.path.exists(out_fis):
    return
  hyp_file_name = meta[model]
  if model == "OPENNMT_BASELINE":
    hyp_file_name = "OPEN_NMT.txt"
  hyp_path = f'{PATH_TO_APPEND}ar-ro/translate/{model}/{prev_dataset}/{hyp_file_name}'
  if model == "human":
    hyp_path = ref_path


  os.system(f"comet-score -s {src_path} "
            f" -t {hyp_path} "
            f"--gpus 1"
            f" --model Unbabel/wmt23-cometkiwi-da-xl > {out_fis}")
```
